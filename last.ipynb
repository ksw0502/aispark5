```python
from sklearn.impute import KNNImputer

def impute_knn(df):
    imputer = KNNImputer(n_neighbors=3)
    df[['PM2.5']] = imputer.fit_transform(df[['PM2.5']])
    df[['기온(°C)']] = imputer.fit_transform(df[['기온(°C)']])
    df[['풍향(deg)']] = imputer.fit_transform(df[['풍향(deg)']])
    df[['풍속(m/s)']] = imputer.fit_transform(df[['풍속(m/s)']])
    df[['강수량(mm)']] = imputer.fit_transform(df[['강수량(mm)']])
    df[['습도(%)']] = imputer.fit_transform(df[['습도(%)']])
    return df
```

```python
import pandas as pd
import numpy as np
from math import radians, sin, cos, asin, sqrt

# Haversine 함수 정의
def haversine(lat1, lon1, lat2, lon2):
    R = 6372.8  # 지구 반지름 (단위: km)
    dLat = radians(lat2 - lat1)
    dLon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)
    a = sin(dLat/2)**2 + cos(lat1)*cos(lat2)*sin(dLon/2)**2
    c = 2*asin(sqrt(a))
    return R*c

# 각 PM2.5 측정소와 가장 가까운 AWS 측정소 찾기
pm_data['aws_lat'] = np.nan
pm_data['aws_lon'] = np.nan
pm_data['distance'] = np.nan
pm_data['aws_location'] = np.nan  # AWS 위치 정보를 저장할 새로운 컬럼 추가
for i, row in pm_data.iterrows():
    lat1, lon1 = row['Latitude'], row['Longitude']
    min_distance = float('inf')
    for j, aws_row in aws_data.iterrows():
        lat2, lon2 = aws_row['Latitude'], aws_row['Longitude']
        distance = haversine(lat1, lon1, lat2, lon2)
        if distance < min_distance:
            min_distance = distance
            pm_data.loc[i, 'aws_lat'] = lat2
            pm_data.loc[i, 'aws_lon'] = lon2
            pm_data.loc[i, 'aws_location'] = aws_row['Location']  # AWS 위치 정보 추가
            pm_data.loc[i, 'distance'] = distance

# 결과 출력
print(pm_data[['Location', 'Latitude', 'Longitude', 'aws_location', 'aws_lat', 'aws_lon', 'distance']])
```

```python
x_train, x_val = X_train[:len(X_train)-216], X_train[len(X_train)-216: ]
y_train, y_val = y_train[:len(y_train)-216], y_train[len(y_train)-216: ]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout,BatchNormalization


batch_size = 32
epochs = 100
### 원래는 0.001 <-성능제일 좋은 비율   #### learning_rate 영향을 많이 받음
learning_rate = 0.001

dropout_rate = 0.01
tf.keras.backend.clear_session()
model = Sequential()
model.add(LSTM(units=16, input_shape=(x_train.shape[1], x_train.shape[2]), return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=32, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=128))
model.add(Dense(units=72, activation='linear'))

optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', min_delta=0.001 ,verbose=1, restore_best_weights=True, patience=5)
# 모델 학습
history = model.fit(x_train, y_train, epochs=epochs,callbacks=[es], batch_size=batch_size, validation_data=(x_val, y_val))
```
