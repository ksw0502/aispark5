```python
from sklearn.impute import KNNImputer

def impute_knn(df):
    imputer = KNNImputer(n_neighbors=7)
    df[['PM2.5']] = imputer.fit_transform(df[['PM2.5']])
    df[['기온(°C)']] = imputer.fit_transform(df[['기온(°C)']])
    df[['풍향(deg)']] = imputer.fit_transform(df[['풍향(deg)']])
    df[['풍속(m/s)']] = imputer.fit_transform(df[['풍속(m/s)']])
    df[['강수량(mm)']] = imputer.fit_transform(df[['강수량(mm)']])
    df[['습도(%)']] = imputer.fit_transform(df[['습도(%)']])
    return df
```

```python
def datetime(num,year,df):
    row=df.loc[df['연도']==num]
    row['일시']= year+'-'+row['일시']
    row['일시']=pd.to_datetime(row['일시'],format='%Y-%m-%d %H:%M')
    df.loc[df['연도']==num,'일시']=row['일시']
```

```python
X_train_list = []
y_train_list = []

for i in range(1, 18):
    df = globals()[f"t{i}"]
    X = []
    y = []
    for j in range(48, len(df)-72):
        X.append(df.iloc[j-48:j, :].values)
        y.append(df.iloc[j:j+72, -1].values)
    X_train_list.append(np.array(X))
    y_train_list.append(np.array(y))

X_train = np.concatenate(X_train_list, axis=0)
y_train = np.concatenate(y_train_list, axis=0)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=RANDOM_SEED)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout,BatchNormalization


batch_size = 32
epochs = 100
### 원래는 0.001 <-성능제일 좋은 비율   #### learning_rate 영향을 많이 받음
learning_rate = 0.001

dropout_rate = 0.01
tf.keras.backend.clear_session()
model = Sequential()
model.add(LSTM(units=16, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=32, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256))
model.add(Dense(units=72, activation='linear'))

optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', min_delta=0.001 ,verbose=1, restore_best_weights=True, patience=5)
# 모델 학습
history = model.fit(X_train, y_train, epochs=epochs,callbacks=[es], batch_size=batch_size, validation_data=(X_val, y_val))
```
